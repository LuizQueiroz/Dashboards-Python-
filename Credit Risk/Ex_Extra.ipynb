```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv('credit_data.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.drop(df[df['age'] < 0].index)
```

```python
df['default'] = [str(i) for i in df['default']]
```

```python
df['age'] = [int(i) if isinstance(i, (float, int)) and not np.isnan(i) else np.nan for i in df['age']]
```

----

### 4. Plotly

```python
px.scatter(df, 'income', 'age')
```

```python
fig = px.scatter(df, x = 'income', y = 'loan', color = 'default', hover_name = 'clientid')
fig
```

```python
fig = px.bar(df, x="default", y="loan", color="default")
fig.show()
```

```python
fig = px.scatter(df, x="income", y="age", size="loan", color="default", log_x = True)
fig.show()
```

### 6. Panel

```python
import panel as pn
pn.extension()
```

```python
int_slider = pn.widgets.EditableIntSlider(name='Idades', start=0, end=100, value=2)
```

```python
int_slider
```

```python
int_slider.value
```

```python
df[df['age'] <= int_slider.value]
```

```python
dfi = df.interactive()
```

```python
df23 = (dfi[dfi['age'] <= int_slider.value])
```

```python
df23.hvplot(kind = 'scatter', x = 'age', y = 'income', color = 'default')
```
